```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Create a figure with three subplots
fig, axes = plt.subplots(1, 3,figsize=(18, 5))




# --- 3. Collective Outlier ---
# Generate time-series data
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.1, 100)

# Introduce a collective outlier (a group of points deviating from the trend)
y[50:60] += 1.5

df_collective = pd.DataFrame({'x': x, 'y': y})

# Plot the data
df_collective.plot(x='x', y='y', ax=axes[2], label='Data')
axes[2].plot(x[50:60], y[50:60], marker='o', linestyle='None', color='red', label='Collective Outlier')

# Add plot features
axes[2].set_title('Collective Outlier Example', fontsize=16)
axes[2].set_xlabel('Time', fontsize=12)
axes[2].set_ylabel('Value', fontsize=12)
axes[2].legend()
axes[2].grid(True)

# Adjust layout and save the plot
plt.tight_layout()
plt.savefig('outlier_types.png')
```
