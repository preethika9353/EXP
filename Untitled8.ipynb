```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/breast-cancer (7).csv')
print(df)

print(df.info())
print(df.isnull().sum())
print(df.describe())
print(df.columns)

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])
print(df['diagnosis'])

x=df.drop('diagnosis',axis=1)
print(x)
y=df['diagnosis']
print(y)

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.23,random_state=0)
print('xtrain shape',xtrain.shape)
print('xtest shape',xtest.shape)
print('ytrain shape',ytrain.shape)
print('ytest shape',ytest.shape)

from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(random_state=1)
model.fit(xtrain,ytrain)
train_accuracy=model.score(xtrain,ytrain)
print(train_accuracy)
test_accuracy=model.score(xtest,ytest)
print(test_accuracy)
train_predictions=model.predict(xtrain)
print(train_predictions)
test_predictions=model.predict(xtest)
print(test_predictions)


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.metrics import mean_squared_error,mean_absolute_error
df=pd.read_csv('/content/housing.csv')
print(df)

df.info()
print(df.isnull().sum())
print(df.head())

plt.scatter(df.built_up_area,df.rent,color='red',marker="*")
plt.xlabel('built_up_area')
plt.ylabel('rent')
plt.title('scatter')
plt.show()

x=df.drop('rent',axis=1)
print(x)
y=df.rent
print(y)

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.33,random_state=42)
print('xtrain shape',xtrain.shape)
print('xtest shape',xtest.shape)
print('ytrain shape',ytrain.shape)
print('ytest shape',ytest.shape)

Reg=linear_model.LinearRegression()
Reg.fit(x,y)
y_pred=Reg.predict(xtest)

Reg.predict([[6,5,15]])
print(Reg.coef_)
print(Reg.intercept_)

z=9.7900215*650*1128.879253617697
print(z)

train_score=Reg.score(xtrain,ytrain)
test_score=Reg.score(xtest,ytest)
print('trainscore(R-squared):',train_score)
print('testscore(R_squared):',test_score)

mse=mean_squared_error(ytest,y_pred)
mae=mean_absolute_error(ytest,y_pred)
print(mse)
print(mae)
```
